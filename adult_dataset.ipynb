```python
# logical order for education levels
education_order = [
    'Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th',
    'HS-grad', 'Some-college', 'Assoc-acdm', 'Assoc-voc', 'Bachelors', 'Masters',
    'Doctorate', 'Prof-school'
]
   
```

```python
 x_train['education'] =x_train['education'].cat.reorder_categories(education_order, ordered=True)
```

```python
#  Create an ordinal encoder using the category dtype with an ordered list
adult['education'] = pd.Categorical(adult['education'], categories=education_order, ordered=True).codes
     
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# Create an instance of OrdinalEncoder with the specified order
ordinal_encoder = OrdinalEncoder(categories=[education_order])
     
```

```python
x_train['education'].head()
```

```python
# Fit and transform the ordinal columns
ordinal_columns=  ['education']
ordinal_encoder.fit(x_train[ordinal_columns])
x_train[ordinal_columns]=ordinal_encoder.transform(x_train[ordinal_columns])
```
