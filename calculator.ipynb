```python
def calculator():
    print("Welcome to the Python Calculator!")

    while True:
        try:
            # Get two values
            num1 = float(input("1:  "))
            num2 = float(input("  "))

            # Show options
           # print("\nChoose operation:")
           # print("1. Addition (+)")
            #print("2. Subtraction (-)")
            #print("3. Multiplication (*)")
            #print("4. Division (/)")
            
            choice = input("Enter choice (+, -, *, /): ")

            if choice == '+':
                result = num1 + num2
            elif choice == '-':
                result = num1 - num2
            elif choice == '*':
                result = num1 * num2
            elif choice == '/':
                if num2 == 0:
                    print("Error: Division by zero is not allowed.")
                    continue
                result = num1 / num2
            else:
                print("Invalid operation. Try again.")
                continue

            print(f"Result: {result}")

        except ValueError:
            print("Invalid input! Please enter numbers.")
            continue

        # Ask if user wants to continue
        repeat = input("\nPress 'C' to calculate again or any other key to exit: ")
        if repeat.upper() != 'C':
            print("Goodbye!")
            break


# Run the calculator
calculator()
```
